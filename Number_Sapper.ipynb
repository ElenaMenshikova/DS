# Number Sapper Game

## Game description
This game was inspired by old good Win Sapper Game and entitled as "Number Sapper Game". The main idea of this game is to guess the numbers which are cells with bombs. The list of these cells are generated randomly at the beggining of the game after the gaming mode choice, and the gamer's task is to input the numbers in the range. If the gamer guess all numbers - it is the win scenario, if he guess at least one - he wooden, which means semi-win scenario, if there are no guessed numbers in a gamer'a set - it is the lose scenario and game is over.
There are 3 gaming modes: light (guess 3 numbers from the 10), normal (guess 4 numbers from the 15) and nightmare (guess 5 numbers from the 20). The gaming mode is made by the gamer at the beginning of the game.

## Action set
1. The gamer choose gaming mode (light, normal, nightmare)
2. Check the gaming mode input:
    if input correct, the game starts
    else, the gamer had disclamer to check input
3. According to the chosen gaming mode, the list of 'cells with the bombs' created as a random numbers set
4. Depending to the gaming mode, gamer input 3-5 integers
5. Check the gamer's number choice
    if correct, the game proceed to the next step
    else, the gamer has disclamer, game terminates
6. Ð¡omparison of the numbers from the 'cells with the bombs' with the gamer input
7. If the gamer guess all numbers - he wins, if he guess at least one - he wooden, if there are no guessed numbers - game is over.
8. The 'cells with the bombs' list of numbers shows to gamer

```python
import random

mode = int(input('Welcome to the new "Number Sapper Game"!\n Your aim \
is to guess and neutralize all the bombs.\n Please, choose your gaming \
mode \n press 1 for light mode,\n press 2 for the normal \
mode,\n press 3 for nightmare mode: '))

# check mode input correctness
def mode_check(mode):
    if mode == 1 or mode == 2 or mode == 3:
        return "Let's start"
    else:
        return "Check mode input"

print(mode_check(mode))

# create mode playground
if mode == 1:
    light_num = random.sample(range(1, 10), 3)
elif mode == 2:
    normal_num = random.sample(range(1, 15), 4)
elif mode == 3:
    mare_num = random.sample(range(1, 20), 5)

# play mechanics
if mode == 1:
    print("Your mode is light.\n There are 10 cells, each of them has unique number from 1 to 10.\
     \n There are 3 bombs placed at random numbers.\n \
    If you guess all numbers - you win, if you guess at least one - you wooden, otherwise - you died.")
    a = int(input('Input the first number: '))
    b = int(input('Input the second number: '))
    c = int(input('Input the third number: '))
    input_light = []
    input_light.append(a)
    input_light.append(b)
    input_light.append(c)
    if len(input_light) != len(set(input_light)):
        print('Input different numbers')
    elif a > 10 or b > 10 or c > 10:
        print('Input number from 1 to 10')
    elif a < 1 or b < 1 or c < 1:
        print('Input number from 1 to 10')
    else:
        if input_light == light_num:
            print('You win!')
        elif a in light_num:
            print('Wounded!')
        elif b in light_num:
            print('Wounded!')
        elif c in light_num:
            print('Wounded!')
        else:
            print('Game over!')
        print('The bombs are in the cells: ', *light_num)
elif mode == 2:
    print("Your mode is normal.\n There are 15 cells, each of them has unique number from 1 to 15.\
         \n There are 4 bombs placed at random numbers.\n \
        If you guess all numbers - you win, if you guess at least one - you wooden, otherwise - you died.")
    a = int(input('Input the first number: '))
    b = int(input('Input the second number: '))
    c = int(input('Input the third number: '))
    d = int(input('Input the fourth number: '))
    input_normal = []
    input_normal.append(a)
    input_normal.append(b)
    input_normal.append(c)
    input_normal.append(d)
    if len(input_normal) != len(set(input_normal)):
        print('Input different numbers')
    elif a > 15 or b > 15 or c > 15 or d > 15:
        print('Input number from 1 to 15')
    elif a < 1 or b < 1 or c < 1 or d < 1:
        print('Input number from 1 to 15')
    else:
        if input_normal == normal_num:
            print('You win!')
        elif a in normal_num:
            print('Wounded!')
        elif b in normal_num:
            print('Wounded!')
        elif c in normal_num:
            print('Wounded!')
        elif d in normal_num:
            print('Wounded!')
        else:
            print('Game over!')
        print('The bombs are in the cells: ', *normal_num)
elif mode == 3:
    print("Your mode is nightmare.\n There are 20 cells, each of them has unique number from 1 to 20.\
             \n There are 5 bombs placed at random numbers.\n \
            If you guess all numbers - you win, if you guess at least one - you wooden, otherwise - you died.")
    a = int(input('Input the first number: '))
    b = int(input('Input the second number: '))
    c = int(input('Input the third number: '))
    d = int(input('Input the fourth number: '))
    e = int(input('Input the fifth number: '))
    input_mare = []
    input_mare.append(a)
    input_mare.append(b)
    input_mare.append(c)
    input_mare.append(d)
    input_mare.append(e)
    if len(input_mare) != len(set(input_mare)):
        print('Input different numbers')
    elif a > 20 or b > 20 or c > 20 or d > 20 or e > 20:
        print('Input number from 1 to 20')
    elif a < 1 or b < 1 or c < 1 or d < 1 or e < 1:
        print('Input number from 1 to 20')
    else:
        if input_mare == mare_num:
            print('You win!')
        elif a in mare_num:
            print('Wounded!')
        elif b in mare_num:
            print('Wounded!')
        elif c in mare_num:
            print('Wounded!')
        elif d in mare_num:
            print('Wounded!')
        elif e in mare_num:
            print('Wounded!')
        else:
            print('Game over!')
        print('The bombs are in the cells: ', *mare_num)
```
