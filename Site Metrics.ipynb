# Metrics for site

There are 3 additional metrics for the the site.
1. Percent of involved users for the period
This indicator might be used for the ad company succcess evaluation. Also we may analyze the ratio of new and regular customers. Also, if we will analyze this metrics during the year, we may realize the top and bottom months for the new customers attraction.
2. Retention
Due to the retention we may find out while customers are get used with our product. To calculate we have to divide the number of unique customers in the previous month to the number of users in the calculated month. For example, in January we had 100 unique customers, in February - 20 unique customers, thus 
ret = 20 / 100 * 100 = 20%
This means that there are 20% of customers who came back for our goods. The good sign is the long-term positive stability.
3. Gross Merchandise Volume (GMV)
Gross Merchandise Volume is the total sum of revenue of the sold goods during the period. This metrics might be also used for compartive goods' evaluation. GMV is calculated as number of purchases * average order value.

```python
# input total revenue
total_revenue = float(input('Input total revenue, $: '))

# input number of customers
num_cust = int(input('Input total number of customers, customers: '))
uniq_cust_jan = int(input('Input number of unique customers in January, customers: '))
uniq_cust_feb = int(input('Input number of unique customers in February, customers: '))

# input number of users
num_user = int(input('Input total number of users number of users who visit a webpage, users: '))
uniq_user_jan = int(input('Input number of unique users who visited a webpage in January, users: '))
uniq_user_feb = int(input('Input number of unique users who visited a webpage in February, users: '))

# input number of purchases
purch = int(input('Input total number of purchases, purchases: '))

# input investment spent
inv_spent = float(input('Input total amount spent for investments, $: '))

# input site / ad metrics
ad_mes_clicks = int(input('Input total amount of clicks on an ad, clicks: '))
ad_imp = int(input('Input total number of times an ad was loaded on a page, impressions: '))
time_user = int(input('Input time measured for each user who visits a webpage, sec: '))
cust_lifespan = int(input('Input total customer lifespan, months: '))
atr_conv = int(input('Inpur total amount of conversion recorded which have been caused clicks, conversion: '))


# check positive numbers
def num(total_revenue, num_cust, uniq_cust_jan, uniq_cust_feb, num_user, uniq_user_jan, uniq_user_feb, purch, inv_spent,
        ad_mes_clicks, ad_imp, time_user, cust_lifespan, atr_conv):
    if total_revenue < 0 or num_cust < 0 or uniq_cust_jan < 0 or uniq_cust_feb < 0 \
    or num_user < 0 or uniq_user_jan < 0 or uniq_user_feb < 0 or purch < 0 or inv_spent < 0\
    or ad_mes_clicks < 0 or ad_imp < 0 or time_user < 0 or cust_lifespan < 0 or atr_conv < 0:
        return 'Input non-negative numbers only'
    else:
        return 'Input correct'


# check user input customers block
def imp_check_cust(num_cust, uniq_cust_jan, uniq_cust_feb):
    if num_cust < uniq_cust_jan or num_cust < uniq_cust_feb:
        return 'Check number of customers'
    elif num_cust < uniq_cust_jan + uniq_cust_feb:
        return 'Check number of customers'
    else:
        return 'Number of customers input correct'


# check user input user block
def imp_check_user(num_user, uniq_user_jan, uniq_user_feb):
    if num_user < uniq_user_jan or num_user < uniq_user_feb:
        return 'Check number of users'
    elif num_user < uniq_user_jan + uniq_user_feb:
        return 'Check number of users'
    else:
        return 'Number of users input correct'


# extremal cases check, zero in division 1
def extremal_cases(ad_imp, inv_spent, num_user, purch, num_cust, ad_mes_clicks):
    if ad_imp == 0 or inv_spent == 0 or num_user == 0 or purch == 0 or num_cust == 0 or ad_mes_clicks == 0:
        return 'Error! Zero in divisor'
    else:
        return 'No zero in divisor'


# extremal cases check, ROI less inv
def roi_less(total_revenue, inv_spent):
    if total_revenue < inv_spent:
        return 'Your ROI is negative'
    else:
        return 'ROI is positive'


# Click-Through Rate (CTR)
def ctr(ad_mes_clicks, ad_imp):
    return (ad_mes_clicks / ad_imp) * 100


# Return on Investment (ROI)
def roi(total_revenue, inv_spent):
    return ((total_revenue - inv_spent) / inv_spent) * 100


# Average Page Time (APT)
def apt(time_user, num_user):
    time_user_1 = time_user
    if time_user < 5:
        time_user_1 = 0
    else:
        time_user = time_user_1
    return time_user / num_user


# Customer Lifetime Value (CLV)
def clv(total_revenue, purch, num_cust, cust_lifespan):
    aver_purch_val = total_revenue / purch
    aver_purch_freq = purch / num_cust
    aver_cust_lifespan = cust_lifespan / num_cust
    return aver_purch_val * aver_purch_freq * aver_cust_lifespan


# check output
print('Processing your input...', '\n',
      num(total_revenue, num_cust, uniq_cust_jan, uniq_cust_feb, num_user, uniq_user_jan,
          uniq_user_feb, purch, inv_spent, ad_mes_clicks, ad_imp, time_user, cust_lifespan, atr_conv), '\n',
      imp_check_cust(num_cust, uniq_cust_jan, uniq_cust_feb),
      '\n', imp_check_user(num_user, uniq_user_jan, uniq_user_feb))

print('Processing your data...', '\n', extremal_cases(ad_imp, inv_spent, num_user, purch, num_cust, ad_mes_clicks),
      '\n',
      roi_less(total_revenue, inv_spent))

if num(total_revenue, num_cust, uniq_cust_jan, uniq_cust_feb, num_user, uniq_user_jan,
       uniq_user_feb, purch, inv_spent, ad_mes_clicks, ad_imp, time_user, cust_lifespan,
       atr_conv) == "Input correct" \
        and imp_check_cust(num_cust, uniq_cust_jan, uniq_cust_feb) == 'Number of customers input correct' \
        and imp_check_user(num_user, uniq_user_jan, uniq_user_feb) == 'Number of users input correct':
    print('Metrics results: ', '\n', 'Click-Through Rate (CTR) = ', ctr(ad_mes_clicks, ad_imp), '\n',
          'Return on Investment (ROI) = ', roi(total_revenue, inv_spent), '\n',
          'Average Page Time (APT) = ', apt(time_user, num_user), '\n', 'Customer Lifetime Value (CLV) = ',
          clv(total_revenue, purch, num_cust, cust_lifespan))
else:
    print('Check your input')

# three more metrics
# Percent involved Users jan, feb IU
def iu(uniq_user_jan, uniq_user_feb, num_user):
    return ((uniq_user_jan + uniq_user_feb) / num_user) * 100


# Retention jan feb
def ret(uniq_user_feb, uniq_user_jan):
    return uniq_user_feb / uniq_user_jan * 100


# Gross Merchandise Volume GMV
def gmv(purch, total_revenue):
    aver_purch_val = total_revenue / purch
    return aver_purch_val * purch


if num(total_revenue, num_cust, uniq_cust_jan, uniq_cust_feb, num_user, uniq_user_jan,
       uniq_user_feb, purch, inv_spent, ad_mes_clicks, ad_imp, time_user, cust_lifespan,
       atr_conv) == "Input correct" \
        and imp_check_cust(num_cust, uniq_cust_jan, uniq_cust_feb) == 'Number of customers input correct' \
        and imp_check_user(num_user, uniq_user_jan, uniq_user_feb) == 'Number of users input correct':
    print('Extra metrics: ', '\n', 'Involved users percent (January - February) = ',
          iu(uniq_user_jan, uniq_user_feb, num_user),
          '\n', 'Retention (January - February) = ', ret(uniq_user_feb, uniq_user_jan), '\n',
          'Gross Merchandise Volume (GMV), $ = ',
          gmv(purch, total_revenue))
else:
    print(' ')

```
